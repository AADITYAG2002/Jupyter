```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sc
```

```python
sample_size = 10000
standard_norm = np.random.normal(size=sample_size)
heavy_tailed_norm = np.random.normal(loc=0, scale=2, size=sample_size)
skewed_norm = sc.skewnorm.rvs(a=5, size=sample_size)
skew_left_norm = sc.skewnorm.rvs(a=-5, size=sample_size)
```

```python
def plot_data(data):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    plt.hist(data)
    plt.subplot(1,2,2)
    sc.probplot(data,dist='norm',plot=plt)
    plt.show()
```

```python
def qq_plot(data):
    fig, ax = plt.subplots(1, 2, figsize=(12, 7))
    sns.histplot(data,kde=True, color ='red',ax=ax[0])
    sc.probplot(data, plot = plt)
```

```python
plot_data(standard_norm)
```

```python
plot_data(heavy_tailed_norm)
```
